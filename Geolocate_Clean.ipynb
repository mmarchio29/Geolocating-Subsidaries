# Creating Geospatial Map

```python
geocoder = OpenCageGeocode('c3e8fc2d7b5e4c318dedf4c22c493da6')

def geocode_address_opencage(row):
    try:
        address_parts = [
            row['Address line 1\nLatin Alphabet'],
            row['Postcode\nLatin Alphabet'],
            row['City\nLatin Alphabet']
        ]
        address = ', '.join(filter(pd.notna, address_parts))

        if address:
            result = geocoder.geocode(address, no_annotations="1")
            if result and len(result):
                return result[0]['geometry']['lat'], result[0]['geometry']['lng'], "Success"
            else:
                return None, None, "No result found"
        else:
            return None, None, "Insufficient address information"
    except Exception as e:
        return None, None, f"Error: {e}"


df_clean[['Latitude', 'Longitude', 'Status']] = df_clean.apply(geocode_address_opencage, axis=1, result_type='expand')


m = folium.Map(location=[45, 0], zoom_start=2)


marker_cluster = MarkerCluster().add_to(m)


plotted_count = 0
not_plotted_count = 0
reasons_for_not_plotting = []

# creating popups
for index, row in df_clean.iterrows():
    if row['Status'] == "Success":
        address_line_1 = row['Address line 1\nLatin Alphabet']
        postcode = row['Postcode\nLatin Alphabet']
        city = row['City\nLatin Alphabet']

        popup_message = f"BvD ID number: {row['BvD ID number']}<br>"
        popup_message += f"Address: {address_line_1}<br>"
        popup_message += f"Postcode: {postcode}<br>"
        popup_message += f"City: {city}"
        popup = folium.Popup(popup_message, max_width=300)

        folium.Marker(
            [row['Latitude'], row['Longitude']],
            popup=popup
        ).add_to(marker_cluster)
        plotted_count += 1
    else:
        not_plotted_count += 1
        reasons_for_not_plotting.append(row['Status'])

# Print the counts and reasons
print("Number of points successfully plotted:", plotted_count)
print("Number of points not plotted:", not_plotted_count)
if reasons_for_not_plotting:
    print("Reasons for not plotting points:")
    for reason in set(reasons_for_not_plotting):
        print(f"{reason}: {reasons_for_not_plotting.count(reason)} times")

# Adding a title to the map
title_html = '''
             <h3 align="center" style="font-size:20px"><b>Company Locations</b></h3>
             '''
m.get_root().html.add_child(folium.Element(title_html))

# displaying the map
m
```

```python
# Saving the map to an HTML file
m.save('company_locations.html')
```

```python
# Function to perform reverse geocoding to get postcode and city
def reverse_geocode_opencage(lat, lng):
    try:
        result = geocoder.reverse_geocode(lat, lng, no_annotations="1")
        if result and len(result):
            # Extracting postcode and city from the result
            components = result[0]['components']
            postcode = components.get('postcode', '').strip()
            city = components.get('city', components.get('town', '')).strip()
            # Return the combined postcode and city only if both are present
            if postcode and city:
                return f"{postcode}, {city}"
        return None
    except Exception as e:
        print(f"Error: {e}")
        return None

# Applying reverse geocoding for rows with successful geolocation
for index, row in df_clean.iterrows():
    if row['Status'] == "Success":
        combined_value = reverse_geocode_opencage(row['Latitude'], row['Longitude'])
        if combined_value:  # Check if the combined value is not None
            df_clean.at[index, 'Postcode_City'] = combined_value
        else:
            df_clean.at[index, 'Postcode_City'] = "Incomplete Data"  # or keep it as None

df_clean.head()  # Displaying the first few rows of the updated DataFrame
```
